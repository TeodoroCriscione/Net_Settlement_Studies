```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
```

```python
resCM = pd.read_csv('./output/results_CM.csv')
resCM.drop(columns=["Unnamed: 0"], inplace=True)
```

```python
df = resCM.loc[resCM.graph_type.isin(["tg_dynet", "tg_statnet"])][["graph_type","synergy","execu_time", "nodes", "edges"]]
fig, axs = plt.subplots(2, 2, figsize=(12, 10))
# first plot
sns.scatterplot(data=df, x='nodes', y='synergy', 
                hue='graph_type', style='graph_type',
                s=100, palette='Set1', ax=axs[0][0])
axs[0][0].set_xlabel('Nodes', fontsize=14, fontweight='bold')
axs[0][0].set_ylabel('Synergy', fontsize=14, fontweight='bold')
axs[0][0].set_xticks(np.arange(100, 1001, 100))
axs[0][0].grid()
axs[0][0].legend_.remove()

# second plot
sns.scatterplot(data=df, x='edges', y='synergy', 
                hue='graph_type', style='graph_type',
                s=100, palette='Set1', ax=axs[0][1])
axs[0][1].set_xlabel('Edges', fontsize=14, fontweight='bold')
axs[0][1].set_ylabel('Synergy', fontsize=14, fontweight='bold')
axs[0][1].set_xticks(list(set(df.edges)))
axs[0][1].tick_params(axis='x', rotation=45) 

# third plot
sns.scatterplot(data=df, x='nodes', y='execu_time', 
                hue='graph_type', style='graph_type',
                s=100, palette='Set1', ax=axs[1][0])
axs[1][0].set_xlabel('Nodes', fontsize=14, fontweight='bold')
axs[1][0].set_ylabel('Exec.Time (seconds)', fontsize=14, fontweight='bold')
axs[1][0].set_xticks(np.arange(100, 1001, 100))
axs[1][0].grid()
axs[1][0].legend_.remove()

# fourth plot
sns.scatterplot(data=df, x='edges', y='execu_time', 
                hue='graph_type', style='graph_type',
                s=100, palette='Set1', ax=axs[1][1])
axs[1][1].set_xlabel('Edges', fontsize=14, fontweight='bold')
axs[1][1].set_ylabel('Exec.Time (seconds)', fontsize=14, fontweight='bold')
axs[1][1].set_xticks(list(set(df.edges)))
axs[1][1].tick_params(axis='x', rotation=45) 
axs[1][1].grid()
axs[1][1].legend_.remove()

# Customize the legend second
handles, labels = axs[0][1].get_legend_handles_labels()
axs[0][1].legend(handles=handles, 
           labels=['Dynamic', 'Static'], 
           title='Netting Algorithms',
           bbox_to_anchor=(1.05, 1),
           loc='upper left',
           borderaxespad=0.,
           fontsize=12,
           title_fontsize=13 )
axs[0][1].grid()
plt.suptitle('Synergy and execution time using dynamic and static algorithms:\nApplication on Configuration Models', fontsize=16, fontweight='bold')
plt.tight_layout()
plt.savefig("./output/CM_results_plot.pdf", format='pdf', bbox_inches='tight')
```
